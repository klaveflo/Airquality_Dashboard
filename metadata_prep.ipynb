```python
# Load the metadata
df = pd.read_csv('station_metadata.csv')

print(f"Original size: {df.shape[0]:,} rows × {df.shape[1]} columns")
print(f"File size: {df.memory_usage(deep=True).sum() / 1024**2:.1f} MB\n")

# Keep only the essential columns for the dashboard
essential_cols = [
    'Country',
    'Air Quality Station Name',
    'Sampling Point Id',
    'Latitude',
    'Longitude',
    'Air Quality Station Area',
    'Operational Activity Begin',
    'Operational Activity End'
]

df_cleaned = df[essential_cols].copy()

# Normalize column names to lowercase with underscores for consistency
df_cleaned.columns = df_cleaned.columns.str.lower().str.replace(' ', '_')

# Deduplicate by Sampling Point Id (keep first occurrence)
# This removes duplicate station entries (same station, different pollutants)
df_cleaned = df_cleaned.drop_duplicates(subset=['sampling_point_id'], keep='first')

print(f"After cleaning:")
print(f"  Rows: {df_cleaned.shape[0]:,} (removed {df.shape[0] - df_cleaned.shape[0]:,} duplicates)")
print(f"  Columns: {df_cleaned.shape[1]}")
print(f"  Memory usage: {df_cleaned.memory_usage(deep=True).sum() / 1024**2:.1f} MB\n")

# Show sample
print("Sample rows:")
print(df_cleaned.head(3))

# Save the cleaned metadata
df_cleaned.to_csv('station_metadata_clean.csv', index=False)
print(f"\n✅ Cleaned metadata saved to 'station_metadata_clean.csv'")
```
